```python
fig = plt.figure(figsize = (10, 5))
p1=plt.bar(df["UNIT"].unique(),df["UNIT"].value_counts(),color=['red','black', 'green', 'blue', 'cyan'],
        width = 0.4)
for i in p1:
  height=i.get_height()
  plt.annotate("{}".format(height),(i.get_x()+i.get_width()/2,height+0.08),ha="center",va="bottom",fontsize=11)
plt.xlabel("UNIT")
plt.ylabel("Count")
plt.title("Fig_1_bar_plot")
plt.show()
```
